```python
import sys
import os
import optstoicpy.script.optstoic as optstoic
from optstoicpy.core import drawpathway
from optstoicpy.core.pathway import Pathway, generate_kegg_model
from optstoicpy.core.database import *
from optstoicpy.script.solver import load_pulp_solver
from optstoicpy.script.optstoic import *
```

## Test Pathway for ketoisovalerate from pyruvate that is 3 reaction steps 

```python
db3 = load_db_v3(
    reduce_model_size=True,
    user_defined_export_rxns_Sji={
        'EX_nad': {'C00003': -1.0},
        'EX_pyruvate': {'C00022': -1.0},
        'EX_nadh': {'C00004': -1.0},
        'EX_h2o': {'C00001': -1.0},
        'EX_hplus': {'C00080': -1.0},
        'EX_nadp': {'C00006': -1.0},
        'EX_nadph': {'C00005': -1.0},
        'EX_co2': {'C00011': -1.0},
        'EX_ketoisv': {'C00141': -1.0}
    }
)
```

```python
pulp_solver = load_pulp_solver(
    solver_names=[
        'SCIP_CMD',
        'GUROBI',
        'GUROBI_CMD',
        'CPLEX_CMD',
        'GLPK_CMD'])
```

```python
custom_flux_constraints = [
    {'constraint_name': 'nadphcons1',
     'reactions': ['EX_nadph', 'EX_nadh'],
     'UB': -1,
     'LB': -1},
    {'constraint_name': 'nadphcons2',
     'reactions': ['EX_nadp', 'EX_nad'],
     'UB': 1,
     'LB': 1},
    {'constraint_name': 'nadphcons3',
     'reactions': ['EX_nadh', 'EX_nad'],
     'UB': 0,
     'LB': 0},
    {'constraint_name': 'nadphcons4',
     'reactions': ['EX_nadph', 'EX_nadp'],
     'UB': 0,
     'LB': 0}]
```

```python
specific_bounds = {'EX_ketoisv': {'LB': 1, 'UB': 1},
                   'EX_pyruvate': {'LB': -2, 'UB': -2},
                   'EX_nadh': {'LB': -1, 'UB': 0},
                   'EX_nad': {'LB': 0, 'UB': 1},
                   'EX_nadph': {'LB': -1, 'UB': 0},
                   'EX_nadp': {'LB': 0, 'UB': 1},
                   'EX_h2o': {'LB': 1, 'UB': 1},
                   'EX_co2': {'LB': 1, 'UB': 1},
                   'EX_hplus': {'LB': -5, 'UB': 5}
                  }  # pulp/gurobi has issue with "h+"
```

```python
test = OptStoic(database=db3,
                objective='MinFlux',
                zlb=None,
                specific_bounds=specific_bounds,
                custom_flux_constraints=custom_flux_constraints,
                add_loopless_constraints=True,
                max_iteration=2,
                pulp_solver=pulp_solver,
                result_filepath='./result1/',
                M=1000)
```

```python
lp_prob, pathways = test.solve()
```

```python
f = open(os.path.join('./result1/test_path.txt'), 'w+')

for ind, p in pathways.items():
    p.rearrange_reaction_order()
    generate_kegg_model(p, filehandle=f)
    graph_title = "{0}_{1}ATP_P{2}".format(p.name, p.nATP, p.id)
    drawpathway.draw_pathway(p,
                             os.path.join('./result1/pathway_{0:03d}'.format(p.id)),
                             imageFormat='png',
                             graphTitle=graph_title)
f.close()
```

## test for glycolysis pathways

```python
db3 = load_db_v3(
    reduce_model_size=True,
    user_defined_export_rxns_Sji={
        'EX_glc': {'C00031': -1.0},
        'EX_nad': {'C00003': -1.0},
        'EX_adp': {'C00008': -1.0},
        'EX_phosphate': {'C00009': -1.0},
        'EX_pyruvate': {'C00022': -1.0},
        'EX_nadh': {'C00004': -1.0},
        'EX_atp': {'C00002': -1.0},
        'EX_h2o': {'C00001': -1.0},
        'EX_hplus': {'C00080': -1.0},
        'EX_nadp': {'C00006': -1.0},
        'EX_nadph': {'C00005': -1.0}
    }
)
```

```python
pulp_solver = load_pulp_solver(
    solver_names=[
        'SCIP_CMD',
        'GUROBI',
        'GUROBI_CMD',
        'CPLEX_CMD',
        'GLPK_CMD'])

custom_flux_constraints = [
    {'constraint_name': 'nadphcons1',
     'reactions': ['EX_nadph', 'EX_nadh'],
     'UB': 2,
     'LB': 2},
    {'constraint_name': 'nadphcons2',
     'reactions': ['EX_nadp', 'EX_nad'],
     'UB': -2,
     'LB': -2},
    {'constraint_name': 'nadphcons3',
     'reactions': ['EX_nadh', 'EX_nad'],
     'UB': 0,
     'LB': 0},
    {'constraint_name': 'nadphcons4',
     'reactions': ['EX_nadph', 'EX_nadp'],
     'UB': 0,
     'LB': 0}]

specific_bounds = {'EX_glc': {'LB': -1, 'UB': -1},
                   'EX_pyruvate': {'LB': 2, 'UB': 2},
                   'EX_nad': {'LB': -2, 'UB': 0},
                   'EX_nadh': {'LB': 0, 'UB': 2},
                   'EX_nadp': {'LB': -2, 'UB': 0},
                   'EX_nadph': {'LB': 0, 'UB': 2},
                   'EX_adp': {'LB': -1, 'UB': -1},
                   'EX_phosphate': {'LB': -1, 'UB': -1},
                   'EX_atp': {'LB': 1, 'UB': 1},
                   'EX_h2o': {'LB': 1, 'UB': 1},
                   'EX_hplus': {'LB': -10, 'UB': 10}}  # pulp/gurobi has issue with "h+"

test_gly = OptStoic(database=db3,
                objective='MinFlux',
                zlb=None,
                specific_bounds=specific_bounds,
                custom_flux_constraints=custom_flux_constraints,
                add_loopless_constraints=True,
                max_iteration=2,
                pulp_solver=pulp_solver,
                result_filepath='./result/',
                M=1000)

lp_prob_gly, pathways_gly = test_gly.solve()


f = open(os.path.join('./result/test_path_gly.txt'), 'w+')

for ind, p in pathways_gly.items():
    p.rearrange_reaction_order()
    generate_kegg_model(p, filehandle=f)
    graph_title = "{0}_{1}ATP_P{2}".format(p.name, p.nATP, p.id)
    drawpathway.draw_pathway(p,
                             os.path.join('./result/pathway_{0:03d}'.format(p.id)),
                             imageFormat='png',
                             graphTitle=graph_title)
f.close()
```
